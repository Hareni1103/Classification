```python
data.drop(['CNT_CHILDREN','FLAG_WORK_PHONE','FLAG_PHONE','FLAG_EMAIL','CNT_FAM_MEMBERS'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in data.select_dtypes(include='object'):
  data[i]=le.fit_transform(data[i])
```

```python
data.head(5)
```

```python
data.columns
```

```python
from sklearn.model_selection import train_test_split
x=data.drop('target',axis=1)
y=data['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

# **BOOSTING**

```python
import xgboost as xgb
xgboost_model=xgb.XGBClassifier()
#train the model
xgboost_model.fit(x_train,y_train)

#make prediction
xgboost_predictions=xgboost_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

xg_accuracy=accuracy_score(y_test,xgboost_predictions)
conf_matrix=confusion_matrix(y_test,xgboost_predictions)
classification_rep=classification_report(y_test,xgboost_predictions)
print(f"accuracy:{xg_accuracy}")
print(f"conf_matrix:{conf_matrix}")
print(f"classification_rep:{classification_rep}")
```

```python
from sklearn.ensemble import AdaBoostClassifier

adamodel=AdaBoostClassifier(n_estimators=50,learning_rate=0.01)
model=adamodel.fit(x_train,y_train)
adamodel_pred=model.predict(x_test)

```

```python
ada_accuracy=accuracy_score(y_test,adamodel_pred)
conf_matrix=confusion_matrix(y_test,adamodel_pred)
classification_rep=classification_report(y_test,adamodel_pred)
print(f"accuracy:{ada_accuracy}")
print(f"conf_matrix:{conf_matrix}")
print(f"classification_rep:{classification_rep}")
```

```python
from catboost import CatBoostClassifier

#create catboost regressor
catboost_model=CatBoostClassifier(iterations=100,depth=6,learning_rate=0.1)

#train the model
catboost_model.fit(x_train,y_train)

#make predictions
catboost_predictions=catboost_model.predict(x_test)
```

```python
cat_accuracy=accuracy_score(y_test,catboost_predictions)
conf_matrix=confusion_matrix(y_test,catboost_predictions)
classification_rep=classification_report(y_test,catboost_predictions)
print(f"accuracy:{cat_accuracy}")
print(f"conf_matrix:{conf_matrix}")
print(f"classification_rep:{classification_rep}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

#create gradient boosting regressor
gradient_boosting_model=GradientBoostingClassifier()

#train the model
gradient_boosting_model.fit(x_train,y_train)

#make predictions
gradient_boosting_predictions=gradient_boosting_model.predict(x_test)
```

```python
gradient_accuracy=accuracy_score(y_test,gradient_boosting_predictions)
conf_matrix=confusion_matrix(y_test,gradient_boosting_predictions)
classification_rep=classification_report(y_test,gradient_boosting_predictions)
print(f"accuracy:{gradient_accuracy}")
print(f"conf_matrix:{conf_matrix}")
print(f"classification_rep:{classification_rep}")
```

# **SUPPORT VECTOR**

```python
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',gamma=0.1,C=1)
classifier.fit(x_train,y_train)

y_pred=classifier.predict(x_test)

```

```python
svc_accuracy=accuracy_score(y_test,y_pred)
conf_matrix=confusion_matrix(y_test,y_pred)
classification_rep=classification_report(y_test,y_pred)
print(f"accuracy:{svc_accuracy}")
print(f"conf_matrix:{conf_matrix}")
print(f"classification_rep:{classification_rep}")
```

# **K-NEAREST NEIGHBORS**

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
model_knn=KNeighborsClassifier(n_neighbors=int(np.sqrt(len(x_train))),metric="euclidean")
model_knn.fit(x_train,y_train)
y_pred_1=model_knn.predict(x_test)
```

```python
knn_accuracy=accuracy_score(y_test,y_pred_1)
conf_matrix=confusion_matrix(y_test,y_pred_1)
classification_rep=classification_report(y_test,y_pred_1)
print(f"accuracy:{knn_accuracy}")
print(f"conf_matrix:{conf_matrix}")
print(f"classification_rep:{classification_rep}")
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
y_pred_3=log_reg.predict(x_test)
```

```python
log_accuracy=accuracy_score(y_test,y_pred_3)
conf_matrix=confusion_matrix(y_test,y_pred_3)
classification_rep=classification_report(y_test,y_pred_3)
print(f"accuracy:{log_accuracy}")
print(f"conf_matrix:{conf_matrix}")
print(f"classification_rep:{classification_rep}")
```
